```python
import pandas as pd
```

```python
df_2013 = pd.read_csv('df_2013.csv')
df_2014 = pd.read_csv('df_2014.csv')
df_2015 = pd.read_csv('df_2015.csv')
df_2016 = pd.read_csv('df_2016.csv')
df_2017 = pd.read_csv('df_2017.csv')
```

```python
df_2013
```

```python
left = df_2013
right = df_2014
df_merge1 = left.merge(right, on=('Unnamed: 0'), suffixes=('_2013', '_2014'))
```

```python
df_merge1
```

```python
left = df_merge1
right = df_2015
# df_merge2 = left.set_index('State').join(right.set_index('State'))
df_merge2 = left.join(right, rsuffix='_2015')
```

```python
df_merge2
```

```python
left = df_merge2
right = df_2016
# df_merge2 = left.set_index('State').join(right.set_index('State'))
df_merge3 = left.join(right, rsuffix='_2016')
```

```python
df_merge3
```

```python
left = df_merge3
right = df_2017
# df_merge2 = left.set_index('State').join(right.set_index('State'))
df_merge4 = left.join(right, rsuffix='_2017')
```

```python
df_merge4
```

```python
median_income = df_merge4.drop(['Unnamed: 0', 'Unnamed: 0_2015', 'Unnamed: 0_2016', 'Unnamed: 0_2017'], axis=1)
```

```python
median_income
```

```python
median_income.to_csv('median_income13_17.csv')
```
